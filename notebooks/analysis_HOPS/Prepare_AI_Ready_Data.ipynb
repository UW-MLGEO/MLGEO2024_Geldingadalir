```python
import obspy
from obspy import read
import numpy as np
import os
```

```python
#Make folder for AI ready data
! mkdir data/ai_ready
filepath = os.getcwd() + '/data/ai_ready/'
clean_filepath = os.getcwd() + '/data/clean/'
```

```python
#read in cleaned mseed files, save as daily npy files and one giant 100 day file
"""
This script processes cleaned mseed files, saves them as daily npy files, and combines them into a single npy file for a specified number of days.

Functions:
    - Reads in cleaned mseed files.
    - Saves each file's data as a daily npy file.
    - Appends each day's data to a cumulative array.
    - Saves the cumulative array as a single npy file representing multiple days of data.

Variables:
    - NUM_FILES (int): The number of mseed files to process.
    - hundred_data (numpy.ndarray): An array to store the cumulative data from all processed files.

Workflow:
    1. Loop through the specified number of files.
    2. Read each mseed file.
    3. Save the data from each file as a daily npy file.
    4. Append the data to the cumulative array.
    5. Save the cumulative array as a single npy file representing all processed days.

Note:
    - The script assumes that the mseed files are named in a sequential manner (e.g., '1_cleaned.mseed', '2_cleaned.mseed', etc.).
    - The script removes an extra point from the data due to decimation before saving.
"""
hundred_data = np.array([])

NUM_FILES = 30  # same variable from data download

for i in range(NUM_FILES):
    print(f'Processing file {i+1} of {NUM_FILES}')
    i += 1

    #read mseed
    st = read(clean_filepath+str(i)+'_cleaned.mseed')

    #save data to day long npys
    data = st[0].data[0:-1] #removing extra point from decimating
    np.save(filepath+str(i)+'_ready.npy', data)

    #append to overall array
    hundred_data = np.append(hundred_data, data)

#save hundred day data npy
np.save(filepath+str(NUM_FILES)+'_days.npy', hundred_data)

print(f'Saved file {NUM_FILES}_days.npy')
```
